### Ej. 1

```python
def get_sample_rate(frequency, duration):
  sample_rate = 22050
  data = librosa.tone(frequency, sr=sample_rate, length=int(sample_rate * duration))
  nyquist = sample_rate/2
  return data, sample_rate, nyquist
```

```python
frequency = 400
duration = 2
data, sample_rate, nyquist = get_sample_rate(frequency, duration)
```

```python
Audio(data=data, rate=sample_rate)
```

### Ej. 2

```python
print(sample_rate)
data_2 = data[::2]
sample_rate_2 = sample_rate//2
print(sample_rate_2)
```

```python
t = np.linspace(0, len(data)/sample_rate, len(data))
t2 = np.linspace(0, len(data_2)/sample_rate_2, len(data_2))
```

```python
plt.figure(figsize=(14, 4))
plt.plot(t, data, '.-', label=f'sr={sample_rate}')
plt.plot(t2, data_2, '.-', label=f'sr={sample_rate_2}')
plt.title(f"Onda senoidal de 400 Hz")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(14, 4))
plt.plot(t, data, '.-', label=f'sr={sample_rate}')
plt.plot(t2, data_2, '.-', label=f'sr={sample_rate_2}')
plt.title(f"Onda senoidal de 400 Hz")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.xlim(0, 0.005)
plt.grid(True)
plt.show()
```

```python
data_reconstructed = np.interp(t, t2, data_2)
```

```python
plt.figure(figsize=(14, 4))
plt.plot(t, data, '.-', label='Original')
plt.plot(t, data_reconstructed, '.-', label='Reconstruido')
plt.title(f"Onda senoidal de 400 Hz")
plt.xlabel("Tiempo [s]")
plt.ylabel("Amplitud")
plt.legend()
plt.xlim(0, 0.005)
plt.grid(True)
plt.show()
```
